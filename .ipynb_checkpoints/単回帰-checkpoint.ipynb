```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pystan
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
df = pd.read_excel('./real_estate1.xlsx', engine='openpyxl')
```

```python
df.head()
```

```python
plt.scatter(df['space'], df['value'])
```

```python
stan_model = """
data {
  int N;
  real X[N];
  real Y[N];
}

parameters {
  real a;
  real b;
  real<lower=0> sigma;
}

model {
  for (n in 1:N){
    Y[n] ~ normal(a * X[n] + b, sigma);
  }
}
"""
```

```python
sm = pystan.StanModel(model_code = stan_model)
```

```python
stan_data = {'N':df.shape[0], 'X':df['space'], 'Y':df['value']}
```

```python
fit = sm.sampling(data = stan_data, iter=2000, warmup=500, chains=3, seed=123)
```

```python
fit
```

```python
fig = fit.plot()
```
